```python
%matplotlib inline
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
from IPython.display import Math, Latex
from IPython.core.display import Image
import seaborn as sns
from scipy.stats import uniform, norm, gamma, expon, poisson, binom, bernoulli
import warnings
warnings.filterwarnings('ignore')

sns.set(color_codes=True)
sns.set(rc={'figure.figsize': (5,5)})
```

### Uniform Distribution

```python
n = 1000
start = 10
width = 20
data_uniform = uniform.rvs(size=n, loc=start, scale=width)

ax = sns.distplot(data_uniform,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={'linewidth': 15, 'alpha': 1})
ax.set(xlabel='Uniform Distribution', ylabel='Freq')
plt.show()
```

### Normal Distribution

```python
n = 10000
start = 0
width = 1
data_norm = norm.rvs(size=n, loc=start, scale=width)
display(data_norm)
print('Mean = {}'.format(np.mean(data_norm)))
print('Std. dev = {}'.format(np.std(data_norm)))

ax = sns.distplot(data_norm,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={'linewidth': 15, 'alpha': 1})
ax.set(xlabel='Normal Distribution', ylabel='Freq')
plt.show()
```

### Gamma Distribution

```python
n = 10000
data_gamma = gamma.rvs(a=5, size=n)
display(data_gamma)
print('Mean = {}'.format(np.mean(data_gamma)))
print('Std. dev = {}'.format(np.std(data_gamma)))

ax = sns.distplot(data_gamma,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={'linewidth': 15, 'alpha': 1})
ax.set(xlabel='Gamma Distribution', ylabel='Freq')
plt.show()
```

### Exponential Distribution

```python
n = 1000
data_exp = expon.rvs(loc=0, size=n, scale=1)
print('Mean = {}'.format(np.mean(data_exp)))
print('Std. dev = {}'.format(np.std(data_exp)))

ax = sns.distplot(data_exp,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={'linewidth': 15, 'alpha': 1})
ax.set(xlabel='Exponential Distribution', ylabel='Freq')
plt.show()
```

### Poisson Distribution

```python
data_poisson = poisson.rvs(mu=3, size=10000)
print('Mean = {}'.format(np.mean(data_poisson)))
print('Std. dev = {}'.format(np.std(data_poisson)))

ax = sns.distplot(data_poisson,
                  bins=30,
                  kde=True,
                  color='skyblue',
                  hist_kws={'linewidth': 15, 'alpha': 1})
ax.set(xlabel='Poisson Distribution', ylabel='Freq')
plt.show()
```

### Binomial Distribution

```python
data_binom = binom.rvs(p=.8, size=10000, n=10)
print('Mean = {}'.format(np.mean(data_binom)))
print('Std. dev = {}'.format(np.std(data_binom)))

ax = sns.distplot(data_binom,
                  bins=30,
                  kde=False,
                  color='skyblue',
                  hist_kws={'linewidth': 15, 'alpha': 1})
ax.set(xlabel='Binomial Distribution', ylabel='Freq')
plt.show()
```

### Bernoulli Distribution

```python
data_bernoulli = bernoulli.rvs(p=.6, size=10000)
print('Mean = {}'.format(np.mean(data_bernoulli)))
print('Std. dev = {}'.format(np.std(data_bernoulli)))

ax = sns.distplot(data_bernoulli,
                  bins=30,
                  kde=False,
                  color='skyblue',
                  hist_kws={'linewidth': 15, 'alpha': 1})
ax.set(xlabel='bernoulliial Distribution', ylabel='Freq')
plt.show()
```

### Hypothesis Testing

```python
import statsmodels.api as sm
```

```python
alpha = .05
n = 1018
pnull = .52
phat = .56
z_calc, pval = sm.stats.proportions_ztest(count=phat*n,
                                          nobs=n,
                                          value=pnull,
                                          alternative='larger')
print('Sig. Level (alpha 5%) = 0.05')
print('P-value = {}'.format(pval))

if pval > alpha:
    print('Accept Null Hypothesis.')
else:
    print('Reject Null Hypothesis.')
```

```python
data = pd.read_csv('nhanes_2015_2016.csv')
data.head()
```

```python
F = data[data['RIAGENDR'] == 2]
M = data[data['RIAGENDR'] == 1]
```

```python
n1 = len(F)
mu1 = F['BMXBMI'].mean()
sd1 = F['BMXBMI'].std()
print('Female --> n={}, mu={}, sd={}'.format(n1, mu1, sd1))
n2 = len(M)
mu2 = M['BMXBMI'].mean()
sd2 = M['BMXBMI'].std()
print('Male --> n={}, mu={}, sd={}'.format(n2, mu2, sd2))
```

```python
z_calc, pval = sm.stats.ztest(x1=F['BMXBMI'].dropna(),
                              x2=M['BMXBMI'].dropna(),
                              alternative='two-sided')
print('Sig. Level (alpha 5%) = 0.05')
print('P-value = {}'.format(pval))

if pval > alpha:
    print('Accept Null Hypothesis.')
else:
    print('Reject Null Hypothesis.')
```
