```python
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd
```

```python
breast = load_breast_cancer()
breast.keys()
```

```python
breast_data = breast.data
print('Features :', breast_data.shape)
print(breast_data)

breast_labels = breast.target
print('Target :', breast_labels.shape)
print(breast_labels)
```

```python
labels = np.reshape(breast_labels, (569, 1))
labels
```

```python
final_breast_data = np.concatenate([breast_data, labels], axis=1)
```

```python
breast_dataset = pd.DataFrame(final_breast_data)
display(breast_dataset.head())

features = breast.feature_names
print(features)
```

```python
pd.set_option('max_columns', None)
```

```python
breast_dataset = pd.DataFrame(columns=list(breast.feature_names)+['label'], data=final_breast_data)
breast_dataset['label'].replace(0, 'Benign', inplace=True)
breast_dataset['label'].replace(1, 'Malignant', inplace=True)
display(breast_dataset.tail())
```

```python
from sklearn.preprocessing import StandardScaler
x = breast_dataset[list(features)].values
x = StandardScaler().fit_transform(x)
print(x.shape)
print(x)
```

```python
normalized_breast = pd.DataFrame(x, columns=list(breast.feature_names))
normalized_breast.head()
```

```python
featurecol_norm = ['feature{}'.format(i) for i in range(len(breast.feature_names))]
normalized_breast = pd.DataFrame(x, columns=featurecol_norm)
normalized_breast.head()
```

```python
from sklearn.decomposition import PCA
pca_breast = PCA(n_components=20)
PCA_breast_result = pca_breast.fit_transform(normalized_breast)
```

```python
pd.DataFrame(PCA_breast_result, columns=['PC{}'.format(i+1) for i in range(20)])
```

```python
pca_breast.explained_variance_ratio_
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
comp = np.arange(pca_breast.n_components_)+1
exp = pca_breast.explained_variance_ratio_
plt.plot(comp, exp, 'o-', linewidth=2, color='red')
plt.title('Scree Plot')
plt.xlabel('PC')
plt.ylabel('Variance Explained')
plt.show()
```

```python
pca_breast = PCA(n_components=2)
PCA_breast_result = pca_breast.fit_transform(normalized_breast)
```

```python
pd.DataFrame(PCA_breast_result, columns=['PC{}'.format(i+1) for i in range(2)])
```

```python
from sklearn.cluster import KMeans
```

```python
X = -2 * np.random.rand(100, 2)
X1 = 1 + 2 * np.random.rand(50, 2)
X[50:100, :] = X1
plt.scatter(X[:, 0], X[:, 1], s=50, c='r')
plt.show()
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, 
                    init='k-means++', 
                    max_iter=300, 
                    n_init=10, 
                    random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('# Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, 
                init='k-means++', 
                max_iter=300, 
                n_init=10, 
                random_state=0)
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
data = pd.read_csv('Wholesale customers data.csv')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
pd.DataFrame(data_scaled).describe()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
SSE = []
for i in range(1, 21):
    kmeans = KMeans(n_clusters=i, n_jobs=-1, init='k-means++')
    kmeans.fit(data_scaled)
    SSE.append(kmeans.inertia_)
    
plt.plot(range(1, 21), SSE, marker='o')
plt.title('The Elbow Method')
plt.xlabel('# Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=6, n_jobs=-1, init='k-means++')
kmeans.fit(data_scaled)
```

```python
prediction = kmeans.predict(data_scaled)
prediction
```

```python
df = pd.DataFrame(data_scaled)
df['cluster'] = prediction
df
```
