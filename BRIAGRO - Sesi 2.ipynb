```python
nama_akun = '@metrotvnews'
target_tweets = 10
tweets_terambil = 0
text_warning = 'ROKOK'

while tweets_terambil != target_tweets:
    tweets_terambil += 1
    tweet = 'ambil tweet ke-{}'.format(tweets_terambil)
    if tweets_terambil == 5:
        tweet = tweet + 'ROKOK'
    print(tweet)
    if text_warning in tweet:
        print('Warning')
        print()
        continue
    
    print('Tweet aman.')
    print()
```
