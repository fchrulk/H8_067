```python
print('Apabila terdapat case penulisan bulan dalam Bahasa Indonesia')
print()
dict_months = {'Januari': 'January',
               'Juni': 'June', 
               'Desember': 'December'}

D = ['2021, 01 Juni', '2021, 13 Desember']
for i in D:
    print('Input -->', i)
    for k, v in dict_months.items():
        if k in i:
            new_i = i.replace(k, v)
            print(new_i)
    print('Output -->', [datetime.strptime(new_i, '%Y, %d %B')])
    print()
    
print('-'*60)
print('Case dalam Bahasa Indonesia di dataset')
dates = pd.date_range(start='2021-01-01', end='2021-12-31', freq='H')
data = pd.DataFrame(dates, columns=['Date'])
data['Date'] = data['Date'].apply(lambda x: x.strftime('%Y, %d %B'))
data['Date'] = data['Date'].str.replace('January', 'Januari')
data['Date'] = data['Date'].str.replace('February', 'Februari')
data['Date'] = data['Date'].str.replace('October', 'Oktober')
data['Date'] = data['Date'].str.replace('December', 'Desember')
print('Before')
display(data)

data['Date'] = data['Date'].str.replace('Januari', 'January')
data['Date'] = data['Date'].str.replace('Februari', 'February')
data['Date'] = data['Date'].str.replace('Oktober', 'October')
data['Date'] = data['Date'].str.replace('Desember', 'December')
print('After')
display(data)
```

### Time series data implementation

```python
opsd_daily = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/opsd_germany_daily.csv')
display(opsd_daily.shape)
display(opsd_daily.head())
display(opsd_daily.tail())
display(opsd_daily.info())

opsd_daily['Date'] = pd.to_datetime(opsd_daily['Date'])
print()
display(opsd_daily.info())
```

```python
min(opsd_daily['Date']), max(opsd_daily['Date'])
```

```python
from tqdm.auto import tqdm
```

```python
check_date_interval = pd.date_range(start=min(opsd_daily['Date']),
                                    end=max(opsd_daily['Date']))
print('Cek apakah banyaknya tanggal sudah sesuai?')
print(len(check_date_interval) == len(opsd_daily))
for d in tqdm(check_date_interval, desc='Cek tanggal'):
    if d not in list(opsd_daily['Date']):
        print('Not in our columns:', d)
```

```python
opsd_daily.set_index('Date',  inplace=True)
display(opsd_daily.tail())
```

```python
opsd_daily = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/opsd_germany_daily.csv',
                         index_col=0,
                         parse_dates=['Date'])
display(opsd_daily.info())
display(opsd_daily.tail())
```

```python
opsd_daily['Year'] = opsd_daily.index.year
opsd_daily['Month'] = opsd_daily.index.month
opsd_daily['Weekday'] = opsd_daily.index.weekday
opsd_daily.tail()
```
