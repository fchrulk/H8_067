```python
df_iceland = df_can.loc['Iceland', years]
display(df_iceland.head())

df_iceland.plot(kind='bar', 
                rot=90,
                figsize=(10,6))

plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Icelandic Immingrants to Canada 1980-2013')

plt.annotate('',
             xy=(32, 70),
             xytext=(28, 20),
             xycoords='data',
             arrowprops=dict(arrowstyle='->', 
                             connectionstyle='arc3',
                             color='blue', 
                             lw=2))

plt.annotate('2008-2011 Financial crisis',
             xy=(28, 30),
             rotation=80,
             va='bottom',
             ha='left')
plt.show()
```

```python
df_pie = df_can.groupby('Continent').sum()[['Total']]
df_pie
```

```python
df_pie['Total']
```

```python
df_pie['Total'].plot(kind='pie', 
            figsize=(5, 6),
            startangle=90,
            shadow=True,
            autopct='%1.5f%%',
           )
plt.title('Immigration to Canada by Continent')
# plt.axis('equal')
plt.show()
```

```python
df_pie['Total'].plot(kind='pie', 
            figsize=(5, 6),
            startangle=90,
            shadow=True,
            autopct='%1.5f%%',
           )
plt.title('Immigration to Canada by Continent')
plt.axis('equal')
plt.show()
```

```python
explod = [.2,0,1,0,0,.5]
df_pie['Total'].plot(kind='pie', 
            figsize=(5, 6),
            startangle=90,
            shadow=True,
            autopct='%1.5f%%',
            explode=explod
           )
plt.title('Immigration to Canada by Continent')
plt.axis('equal')
plt.show()
```
