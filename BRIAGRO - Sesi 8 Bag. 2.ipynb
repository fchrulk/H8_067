```python
latitude = 37.773972
longitude = -122.431297

san_fracisco_map = folium.Map(location=[latitude, longitude], zoom_start=12)

incidents = folium.map.FeatureGroup()

for lat, long, cat in zip(df_incidents['Y'], df_incidents['X'], df_incidents['Category']):
    if cat not in ['OTHER OFFENSES', 'NON-CRIMINAL']:
        incidents.add_child(
            folium.CircleMarker(
                [lat, long],
                radius=4,
                color=None,
                popup=cat,
                fill=True,
                fill_color='blue',
                fill_opacity=.6
            )
        )
    elif cat == 'OTHER OFFENSES':
        incidents.add_child(
            folium.CircleMarker(
                [lat, long],
                radius=4,
                color=None,
                popup=cat,
                fill=True,
                fill_color='red',
                fill_opacity=.6
            )
        )
    else:
        incidents.add_child(
            folium.CircleMarker(
                [lat, long],
                radius=4,
                color=None,
                popup=cat,
                fill=True,
                fill_color='green',
                fill_opacity=.6
            )
        )

# latitudes = list(df_incidents['Y'])
# longitudes = list(df_incidents['X'])
# labels = list(df_incidents['Category'])

# for lat, long, label in zip(latitudes, longitudes, labels):
#     folium.Marker([lat, long], popup=label).add_to(san_fracisco_map)
    
san_fracisco_map.add_child(incidents)
```

```python
df_can = pd.read_excel('Canada.xlsx', 
                       sheet_name='Canada by Citizenship', 
                       skiprows=range(20), 
                       skipfooter=2)
df_can.drop(columns=['AREA', 'REG', 'DEV', 'Type', 'Coverage'], inplace=True)
df_can.rename(columns={'OdName': 'Country',
                       'AreaName': 'Continent',
                       'RegName': 'Region'}, inplace=True)
df_can.columns = list(map(str, df_can.columns))
df_can['Total'] = df_can.sum(axis=1)
years = list(map(str, range(1980, 2014)))
df_can.head()
```
